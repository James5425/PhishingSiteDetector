```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Clean / Load Data

# Load data
df = pd.read_csv('Data/Phishing_Data.csv')
```

```python
# Change -1 to 0

df['Result'] = df['Result'].replace(-1, 0)
```

```python
# Data Exploration

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Heatmap

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# Visualizing the target variable

df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# Assigning features and target variable
target = df['Result']
features = df.drop('Result', axis=1)
```

```python
# Splitting the data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
pred = logreg.predict(X_test)
```

```python
print("Accuracy of Logistic Regression: ", accuracy_score(y_test, pred))
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

cmLog = confusion_matrix(y_test, pred)
print(cmLog)

classificLog = classification_report(y_test, pred)
print(classificLog)
```

```python
# Feature Importance

plt.figure(figsize=(12, 8))
n_features = X_train.shape[1]
plt.barh(range(n_features), logreg.coef_[0], align='center')
plt.yticks(np.arange(n_features), features.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
pred = dt.predict(X_test)
```

```python
print("Accuracy of Decision Tree: ", accuracy_score(y_test, pred))
```

```python
#Confusion matrix
cmLog = confusion_matrix(y_test, pred)
print(cmLog)

classificLog = classification_report(y_test, pred)
print(classificLog)
```

```python
# Feature Importance

plt.figure(figsize=(12, 8))
n_features = X_train.shape[1]
plt.barh(range(n_features), dt.feature_importances_, align='center')
plt.yticks(np.arange(n_features), features.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_test)
```

```python
print("Accuracy of Random Forest: ", accuracy_score(y_test, pred))
```
